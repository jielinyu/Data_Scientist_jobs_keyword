```python

import pandas as pd
import numpy as np
import time
from urllib.request import urlopen
import requests
from graphviz import Digraph
import re
import time
import warnings
warnings.filterwarnings("ignore")

# Web Scraping 
from bs4 import BeautifulSoup
import spacy

#nltk
import nltk
from nltk import bigrams, trigrams
from nltk import ngrams
from nltk.tokenize import word_tokenize

# Visualization 
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

# sklearn 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn import cluster
from sklearn import metrics
from collections import Counter
from sklearn.cluster import DBSCAN

```

### Web Scraping 200 Data Scientist Jobs from Indeed

```python
def web_scrape(search_position,search_location):
    """
    Given desired position and location, return 200 jobs information table and save it to csv file. 
    
    Keyword arguments:
    search_position -- (str) the jobs position 
    search_location -- (str) the jobs location
    
    Returns: Head of table
    """
    job_list = []
    title = []
    location = []
    company = []
    description = []
    url_http = "https://www.indeed.ca/m/"
    # Scrape 200 jobs 
    for i in range(20):
        time.sleep(1)
        jobs_html = requests.get("https://www.indeed.ca/m/"+"jobs?q="+search_position+"&l="+search_location+"&start="+str(i*10))
        job_bs = BeautifulSoup(jobs_html.text)
        for link in job_bs.find_all('a',{'rel':"nofollow"}):
            job_list.append(url_http + link.get('href'))
    # Scrape jobs' information       
    for job in job_list:
        job_50 = requests.get(job)
        bs_50 = BeautifulSoup(job_50.text)
        title.append(bs_50.find('font',{'size':'+1'}).text)
        location.append(bs_50.find('span',{'class':'location'}).text)
        company.append(bs_50.find('span', {'class':"location"}).previous_sibling)
        description.append(bs_50.find('div',{'id':'desc'}).text)
    # Make a dataframe 
    table = pd.DataFrame({'Company':company,
              'Location':location,
              'Title':title,
              'Description':description,
              'Link':job_list})
    # Save csv
    table.to_csv('Data_Scientist_jobs.csv')
    
    return table.head()
```

```python
search_position = "data+scientist"
search_location = "Canada"
web_scrape(search_position,search_location)
```

### Clean and Preprocess Job Description

```python
nlp = spacy.load("en")
def preprocess(text,  
               irrelevant_pos = ['ADV','PRON','CCONJ','PUNCT','PART','DET','ADP','SPACE','VERB']): 
    """
    Given text, and irrelevant_pos carry out preprocessing of the text 
    and return a preprocessed string. 
    
    Keyword arguments:
    text -- (str) the text to be preprocessed
    irrelevant_pos -- (list) a list of irrelevant pos tags
    
    Returns: (str) the preprocessed text
    """
    # Remove distracting characters
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'''[\*\~]+''', "", text)
    
    # Customized Stopwords 
    # The job postings contains some French as these are jobs in Canada
    stopwords = [
    "jobs","apply","email","password","new","please","required",
     "opportunities","skills","sign","talent","follow",
     "job","work","click","home","next","us","start","take","enter"
     ,"date","people","career","hours","week","type","attach","resume","search","advanced","privacy","terms",
     "find","years","experience","find","title","keywords","company","indeed","help","centre","cookies",
     "post","(free)","industry","employment","ago","save","forums","browse","employer","browse","city",
     "province","title","api","review","application","help","instructions","applying","full","time","opportunity",
     "team","join","resumes","employers","rights","reserved","create","north","america","toronto","interview","cover","letter",
     "30+","days","inc","trends",'de','le','ericsson','dans','des','working','etc.','canada','les','employees','accommodation',
        'et', 'montréal','en','mi','yoga','à','équipe','vancuover','se','se se' 
    ]
    
    text = " ".join([i.lower() for i in text.split() if i.lower() not in stopwords])
    
    doc = nlp(text)
    clean_text = []
    
    for token in doc:
        if (token.is_stop == False # Check if it's not a stopword
            and token.is_alpha # Check if it's an alphanumerics char 
            and token.pos_ not in irrelevant_pos): # Check if it's irrelevant pos
            clean_text.append(str(token))
    return " ".join(clean_text)
```

```python
table['clean_description'] = table['Description'].apply(preprocess)
```

## Method 1: Most 20 Common n-gram words 

```python
# Show frequency of contiguous sequence of 1,2,3 items respectively 
sent_list = [i.split() for i in table['clean_description'].tolist()]
# Unigram
unigram = [[gram for gram in ngrams(sent,1)] for sent in sent_list]
uni_20 = Counter(sum(unigram, [])).most_common(20)
# Bigrams
bigrams = [[gram for gram in ngrams(sent,2)] for sent in sent_list]
bi_20 = Counter(sum(bigrams, [])).most_common(20)
#Trigrams
trigrams = [[gram for gram in ngrams(sent,3)] for sent in sent_list]
tri_20 = Counter(sum(trigrams, [])).most_common(20)

pd.DataFrame({'Unigram':uni_20,'Bigram':bi_20,'Trigram':tri_20})
```

### Visualization

Unigram does not really give us meaningful insights, let's visualize Bigram and Trigram words count. 

```python
bi = [bi_20[i][0] for i in range(len(bi_20))]
height_bi = [bi_20[i][1] for i in range(len(bi_20))]
y_pos = np.arange(len(bi))
tri = [tri_20[i][0] for i in range(len(tri_20))]
height_tri = [tri_20[i][1] for i in range(len(tri_20))]
y_pos_tri = np.arange(len(tri))
# Plot Bigram 
plt.figure(figsize=(12,10))
plt.subplot(2,1,1)
plt.barh(y_pos, height_bi)
plt.yticks(y_pos, bi)
plt.title("Bigram Count")
plt.gca().invert_yaxis();
# Plot Trigram 
plt.subplot(2,1,2)
plt.barh(y_pos_tri, height_tri)
plt.yticks(y_pos_tri, tri)
plt.title("Trigram Count")
plt.gca().invert_yaxis();
```

From the plot above, we can see that **Machine Learning**, **Computer Science**, **Data Mining**, **Deep Learning**, **Python**, **R**, **Big Data**, and **Artificical Intelligence** are important skills for data scientist since these key words are mentioned most frequently in most of the job postings. **Machine Learning** seems is the most important skills for data scientist. 

## Method 2: TF-IDF Vectorizer 

The TF-IDF algorithm is used to weigh a keyword in any content and assign the importance to that keyword based on the number of times it appears in the document. More importantly, it checks how relevant the keyword is throughout the web, which is referred to as corpus.
- TF refers to term frequency of a term in a document.
- DF refers to the number of documents of the corpus in which this term occurs.

```python
vectorizer = TfidfVectorizer(use_idf=True, 
                             tokenizer=nltk.word_tokenize, 
                             ngram_range=(2, 5),max_features=100)
tfidf_vectors = vectorizer.fit_transform(table['clean_description'])
tfidf_post = pd.DataFrame(tfidf_vectors.toarray(), 
                          columns=vectorizer.get_feature_names())
plt.figure(figsize=(12,6))
# Take average of tfidf vectors (TF/DF)
tfidf_post.mean(axis=0).sort_values(ascending=False)[:30].plot.barh()
plt.gca().invert_yaxis();
```

From the plot above, the critial skills extracting from TF-IDF vectorizer are pretty similar to the skills extracting from N-grams. In addition to the technical skills I find above, this method also indicates **soft skills** are important, including verbal communication, ability to solve business problem and **education** bachelor degree. It aligns with the findings above that machine learning is the most important skill.

## Clustering - Kmeans

I think **Kmeans** is the best clustering algorithm to be used in order to extract key skills from job postings. Since we do not have labelled data, I can only use unsupervised learning, which includes Kmeans, DBScan, hierarchical clustering, and LDA model. Kmeans is one of the most effective clustering algorithm for text data. Kmeans algorithm is an iterative algorithm that tries to partition the dataset into K pre-defined distinct non-overlapping subgroups (clusters) where each data point belongs to only one group. It tries to make the inter-cluster data points as similar as possible while also keeping the clusters as different (far) as possible.

To choose the number of clusters, I use **Elbow Method**. Elbow method gives us an idea on what a good k number of clusters would be based on the sum of squared distance (SSE) between data points and their assigned clusters’ centroids. We pick k at the spot where SSE starts to flatten out and forming an elbow. 

```python
wcss = []
for i in range(1,8):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(tfidf_vectors)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,8),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
```

From the plot, it is not obvious when SSE forms an elbow. Therefore, let's try 2, 3, and 4 clusters. 

```python
for ncluster in range(2,5):
    print("Number of Clusters:" + str(ncluster))
    words = vectorizer.get_feature_names()
    kmeans = KMeans(n_clusters = ncluster, n_init = 10, n_jobs = 1,init='k-means++',random_state=42) # n_init(number of iterations for clsutering) n_jobs(number of cpu cores to use)
    kmeans.fit(tfidf_vectors)
# We look at 3 the clusters generated by k-means.
    common_words = kmeans.cluster_centers_.argsort()[:,-1:-16:-1]
    for num, centroid in enumerate(common_words):
        print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
        print("-----------------")
    #print("----------------------------------------------------------------")
```

## 3 Clusters

```python
words = vectorizer.get_feature_names()
kmeans = KMeans(n_clusters = 3, n_init = 10, n_jobs = 1,init='k-means++',random_state=42) # n_init(number of iterations for clsutering) n_jobs(number of cpu cores to use)
kmeans.fit(tfidf_vectors)
# We look at 3 the clusters generated by k-means.
common_words = kmeans.cluster_centers_.argsort()[:,-1:-16:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
    print("-----------------")
```

From the results above, I think clustering skill sets to 3 groups is the best way as I can see clear distinction among three groups. We can clearly see that the first critical skill set is about **soft skills**, including verbal communication, public service, research development, and canadaian citizen; the second skill set is **cutting-edge technical skills**, including machine learning, AI, NLP, deep learning and data mining; the third skill set is **techinical tools**, including R, Python, open source and large data sets.
